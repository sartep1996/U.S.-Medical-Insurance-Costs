```python
import csv

column_index = 5
column_data = []

with open('insurance.csv', 'r') as file:
    insurance_regions = csv.reader(file)
    for row in insurance_regions:
        if len(row) > column_index:
            column_data.append(row[column_index])

for data in column_data:
    print(data)
```

```python
from collections import Counter

most_common = Counter(column_data)

most_common_item = most_common.most_common()

print(most_common_item)
```

```python
most_common_item = None
most_common_count = 0

item_counts = {}

for item in column_data:
    if item in item_counts:
        item_counts[item] += 1
    else:
        item_counts[item] = 1

    if item_counts[item] > most_common_count:
        most_common_item = item
        most_common_count = item_counts[item]

print("Most common item:", most_common_item)
print("Count:", most_common_count)
```

```python
column_index = 4
column_index_2 = 6

smokers = []
spendings = []

with open('insurance.csv', 'r') as file:
    insurance = csv.reader(file)
    for row in insurance:
        if len(row) > column_index:
            smokers.append(row[column_index])

        if len(row) > column_index_2:
            spendings.append(row[column_index_2])

spendings.pop(0)
spendings_int = []
for num in spendings:
    spendings_int.append(float(num))

spendings = spendings_int

print(smokers)
print(spendings)
```

```python
smoker_costs = list(zip(smokers, spendings))
print(smoker_costs)
```

```python
yes = 0
no = 0

for condition in smoker_costs:
    if condition[0] == 'yes':
        yes += condition[1]
    else:
        no += condition[1]

print("Average cost for smokers is: " + str(yes/len(smoker_costs)))
print("Average cost for non smokers is: " + str(no/len(smoker_costs)))
```

```python
column_index = 0
column_index_2 = 3

age = []
children = []

with open('insurance.csv', 'r') as file:
    insurance = csv.reader(file)
    for row in insurance:
        if len(row) > column_index:
            age.append(row[column_index])

        if len(row) > column_index_2:
            children.append(row[column_index_2])

del age[0], children[0]

age_int = [int(x) for x in age]
children_int  = [int(x) for x in children]

age, children = age_int, children_int


print(age)
print(children)
```
